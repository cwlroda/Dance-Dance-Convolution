```python
# import cv2
import time
import random

frame = cv2.imread("/Users/cwlroda/Downloads/lel.jpeg", 1)

start = time.time()
stimer = time.time()

numframes = 100

database = ["T", "V", "I"]
defaultscore = 1
totalscore = 0
maxtime = 5

for _ in range(numframes):
    time.sleep(random.randint(0,5))
    end = time.time()
    elapsed = int(end - start)
    
    if end - stimer >= 10:
        break
    
    shape = random.choice(database)
    angle = random.randint(0,101)
    
    print("Shape: {}".format(shape))
    print("Angle: {}".format(angle))
    print("Time: {}".format(elapsed))
    start = time.time()
    
    score = 0

    if shape == "T":
        if angle > -10 and angle < 15:
            score = (maxtime - elapsed) * defaultscore
        else:
            score = 0

    elif shape == "V":
        if angle > 30 and angle < 60:
            score = (maxtime - elapsed) * defaultscore
        else:
            score = 0

    elif shape == "I":
        if angle > 75 and angle < 100:
            score = (maxtime - elapsed) * defaultscore
        else:
            score = 0
    
    if score == 0:
        print("Try again")
    
    elif score == 1:
        print("Just made it")
        
    elif score == 2:    
        print("Close call")
        
    elif score == 3:   
        print("Good")
        
    elif score == 4:    
        print("Great")
        
    elif score == 5:    
        print("Perfect")
    
    totalscore += score
    score = 0

    print("Total Score: {}\n".format(totalscore))
    
print("Time's up")

font = cv2.FONT_HERSHEY_SIMPLEX

cv2.putText(frame, "Your total score is: {}".format(totalscore), (10,500), font, 2, (255,255,255), 10)
cv2.imshow("DDC", frame)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
