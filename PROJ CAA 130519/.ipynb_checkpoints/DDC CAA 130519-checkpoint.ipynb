# Defining Functions
Determining left and right inputs:
Case 0: no input
Case 1: 0 deg input
Case 2: 45 deg input
Case 3: 90 deg input
Case 4: -45 deg output


Determining final inputs:
Case 0: Invalid
Case 1:  L = 0   R = 0
Case 2:  L = 0   R = 45
Case 3:  L = 0   R = 90
Case 4:  L = 0   R = -45
Case 5:  L = 45  R = 0
Case 6:  L = 45  R = 45
Case 7:  L = 45  R = 90
Case 8:  L = 45  R = -45
Case 9:  L = 90  R = 0
Case 10: L = 90  R = 45
Case 11: L = 90  R = 90
Case 12: L = 90  R = -45
Case 13: L = -45 R = 0
Case 14: L = -45 R = 45
Case 15: L = -45 R = 90
Case 16: L = -45 R = -45

```python
def determineleftinput(leftangle, leftheight):
    if leftangle > -10 and leftangle < 10 and leftheight < 3:
        return 0
    elif leftangle > -10 and leftangle < 10 and leftheight > 3:
        return 1
    elif leftangle > 30 and leftangle < 60:
        return 2
    elif leftangle < -80 or leftangle > 80:
        return 3
    elif leftangle > -60 and leftangle < -30:
        return 4
    
def determinerightinput(rightangle, rightheight):
    if rightangle > -10 and rightangle <10 and rightheight <3:
        return 0
    elif rightangle > -10 and rightangle  < 10 and rightheight > 3:
        return 1
    elif rightangle >30 and rightangle < 60:
        return 2
    elif rightangle <-80 or rightangle >80:
        return 3
    elif rightangle >-60 and rightangle <-30:
        return 4
    
def determinefinalinput(leftangle, rightangle, leftheight, rightheight):
    left = determineleftinput(leftangle, leftheight)
    right = determinerightinput(rightangle, rightheight)
    if left == 0 or right == 0:
        return 0
    elif left == 1 and right == 1:
        return 1
    elif left == 1 and right == 2:
        return 2
    elif left == 1 and right == 3:
        return 3
    elif left == 1 and right == 4:
        return 4
    elif left == 2 and right == 1:
        return 5
    elif left == 2 and right == 2:
        return 6
    elif left == 2 and right == 3:
        return 7
    elif left == 2 and right == 4:
        return 8
    elif left == 3 and right == 1:
        return 9
    elif left == 3 and right == 2:
        return 10
    elif left == 3 and right == 3:
        return 11
    elif left == 3 and right == 4:
        return 12
    elif left == 4 and right == 1:
        return 13
    elif left == 4 and right == 2:
        return 14
    elif left == 4 and right == 3:
        return 15
    elif left == 4 and right == 4:
        return 16
```
